```python
import librosa as lbrs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

%matplotlib inline
sns.set()
```

```python
happy = 'pharrell_williams_happy.mp3'
file_path = sys.path[0] + '/Data/' + happy
```

```python
waveform, sr = lbrs.load(file_path, sr= None)
```

```python
x = range(waveform.size)
plt.plot(x, waveform)
plt.show()
```

```python
tempo, beat_frames = lbrs.beat.beat_track(waveform, sr=sr) #actually 156 bpm
```

```python
beat_times = lbrs.frames_to_time(beat_frames, sr=sr)
```

```python
wf_harmonics, wf_percussive = lbrs.effects.hpss(waveform)
```

```python
fig, ax_list = plt.subplots(2,1)
ax_list[0].plot(x, wf_harmonics)
ax_list[1].plot(x, wf_percussive)
plt.show()
```
